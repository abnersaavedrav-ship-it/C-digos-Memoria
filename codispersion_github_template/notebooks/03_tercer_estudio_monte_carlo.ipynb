Este notebook ejecuta el loop Monte Carlo con CBB pareado para una grilla de escenarios `(rho0, range_pix, anis_mode)`. La configuración inicial es pequeña para validar el flujo.

```python
# Ejecutar una vez si aparece ModuleNotFoundError
from pathlib import Path
import sys, subprocess
cwd = Path.cwd().resolve()
candidates = [cwd, cwd.parent, cwd.parent.parent]
ROOT = None
for path in candidates:
    if (path / "pyproject.toml").exists() and (path / "src" / "codispersion_bootstrap").exists():
        ROOT = path
        break
if ROOT is None:
    raise RuntimeError(f"No encontré la carpeta raíz del proyecto. Estoy parado en: {cwd}")
print("Carpeta raíz encontrada:", ROOT)
subprocess.check_call([sys.executable, "-m", "pip", "install", "-e", str(ROOT)])
print("Paquete instalado correctamente. Reinicia el kernel después de esta celda si es primera vez.")
```

```python
from codispersion_bootstrap import build_H, run_monte_carlo_matern_lmc, monte_carlo_to_dataframe
```

```python
H = build_H(hmax=2)
res = run_monte_carlo_matern_lmc(
    n1=128, n2=128,         # más pequeño para test
    H=H,
    rhos=[0.5],             # 1 valor de rho0
    ranges=[7.0],           # 1 rango
    anis_modes=["aniso"],   # 1 modo
    b=41, B=120,            # boot pequeño para test
    R=10,                   # 10 réplicas Monte Carlo (test)
    nu=1.0,
    angle_deg=45.0,
    anis_ratio=2.0,
        base_seed=2025,
)
df = monte_carlo_to_dataframe(res)
df
```
