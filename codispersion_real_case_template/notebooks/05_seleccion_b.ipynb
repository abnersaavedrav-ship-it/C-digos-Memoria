# 5. Selección de b

Incluye selección rápida por score y opción double bootstrap.

```python
# Ejecutar una vez si aparece ModuleNotFoundError.
from pathlib import Path
import sys, subprocess
cwd = Path.cwd().resolve()
candidates = [cwd, cwd.parent, cwd.parent.parent]
ROOT = None
for p in candidates:
    if (p / "pyproject.toml").exists() and (p / "src" / "codispersion_real_case").exists():
        ROOT = p
        break
if ROOT is None:
    raise RuntimeError(f"No encontré la raíz del proyecto desde {cwd}")
subprocess.check_call([sys.executable, "-m", "pip", "install", "-e", str(ROOT)])
print("Proyecto instalado desde:", ROOT)
```
